#### Problema - O nome das pessoas tem vírgula - REGEX PAL .* (qualquer texto qualquer quantidade) \s -> espaço em regex

#### https://www.regexpal.com/

## O objetivo é retirar a vírgula e inverter a posição dos nomes

```python
import re

match = re.compile(r'"(.*)(,)(\s.*)"')
match.sub(r'"\3 \1"', linhas[1])
```

### Criando a função que remove a vírgula dos nomes e troca a posição

```python
def tratar_nome(linha):
    import re
    match = re.compile(r'"(.*)(,)(\s.*)"')
    return match.sub(r'"\3 \1"', linha)
```

```python
# Vamos testar a função na linha 1
tratar_nome(linhas[1])
```

```python
linha1_tratada = tratar_nome(linhas[1])
type(linha1_tratada)
```

```python
linha1_separada = linha1_tratada.split(',')
```

```python
print(colunas)
print(linha1_tratada)
print(linha1_separada)
```

```python
# Qual a diferença da tratada e a separada?
print(type(linha1_tratada))
print(type(linha1_separada))
```

```python
print(len(colunas), len(linha1_separada))
```

### Criando o dicionário - o dicionário titanic_dados está recebendo a coluna como chave e uma lista vazia como valores

```python
# Na criação desse dicionário, criaremos as chaves e os valores como vazios
titanic_dados = {}

for coluna in colunas:
    titanic_dados[coluna] = []
titanic_dados
```

```python
# Aqui trataremos todas as linhas, excluindo a linha 0, que é o cabeçalho
for dado in linhas[1:]:
    dado_tratado = tratar_nome(dado)
    print(dado_tratado)
```

```python
# Aqui faremos a mesma coisa, porém vamos separar as linhas pela vírgula e converter em uma lista!
for dado in linhas[1:]:
    dado_tratado = tratar_nome(dado)
    dado_como_lista = dado_tratado.split(',')
    print(dado_como_lista)
```

```python
dado_como_lista[0]
```

```python
# Aqui faremos a mesma coisa dos passos anteriores, porém vamos criar uma lista para adicionar cada linha convertida em
# lista, fazendo uma lista aninhada.
dados_tratados = []

for dado in linhas[1:]:
    dado_tratado = tratar_nome(dado)
    dado_como_lista = dado_tratado.split(',')
    dados_tratados.append(dado_como_lista)
```

### Utilizando o ENUMERATE. O enumerate retorna o ÍNDICE em que o elemento está e também o seu valor!

```python
frutas = ['mamão', 'melão', 'melancia']

for fruta in frutas:
    print(fruta)
```

```python
frutas = ['mamão', 'melão', 'melancia']

for indice, fruta in enumerate(frutas):
    print('Fruta:', fruta, 'no índice', indice)
```

```python
frutas = ['mamão', 'melão', 'melancia']

for indice, fruta in enumerate(frutas):
    print(fruta, indice)
```

```python
# 1 - Pego cada linha da variável dados_tratados e coloco em passageiro
# 2 - Faço um for em passageiro e pego o índice e o restante dos dados
# 3 - Vou na variável coluna e pego a coluna no respectivo índice
for passageiro in dados_tratados:
    for indice, dado in enumerate(passageiro):  # Enumerate retorna o ÍNDICE e VALOR (atribuindo MÚLTIPLA)
        coluna = colunas[indice]
        print(coluna, dado)
```

```python
lista_idades = []

for idade in titanic_dados['Age']:
    if idade != -1:
        lista_idades.append(idade)
```

```python
def media(lista):
    soma_total = sum(lista)
    quantidade = len(lista)
    return soma_total / quantidade
```

```python
print(media(lista_idades))
```

```python
media_idades = media(lista_idades)
```

```python
for indice, age in enumerate(titanic_dados['Age']):
    if age == 1:
        titanic_dados['Age'][indice] = media_idades
```
