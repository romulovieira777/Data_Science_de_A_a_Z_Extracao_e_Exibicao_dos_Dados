#### Problema - O nome das pessoas tem vírgula - REGEX PAL .* (qualquer texto qualquer quantidade) \s -> espaço em regex

#### https://www.regexpal.com/

## O objetivo é retirar a vírgula e inverter a posição dos nomes

```python
import re

match = re.compile(r'"(.*)(,)(\s.*)"')
match.sub(r'"\3 \1"', linhas[1])
```

### Criando a função que remove a vírgula dos nomes e troca a posição

```python
def tratar_nome(linha):
    import re
    match = re.compile(r'"(.*)(,)(\s.*)"')
    return match.sub(r'"\3 \1"', linha)
```

```python
# Vamos testar a função na linha 1
tratar_nome(linhas[1])
```

```python
linha1_tratada = tratar_nome(linhas[1])
type(linha1_tratada)
```

```python
linha1_separada = linha1_tratada.split(',')
```

```python
print(colunas)
print(linha1_tratada)
print(linha1_separada)
```

```python
# Qual a diferença da tratada e a separada?
print(type(linha1_tratada))
print(type(linha1_separada))
```

```python
print(len(colunas), len(linha1_separada))
```

### Criando o dicionário - o dicionário titanic_dados está recebendo a coluna como chave e uma lista vazia como valores

```python
# Na criação desse dicionário, criaremos as chaves e os valores como vazios
titanic_dados = {}

for coluna in colunas:
    titanic_dados[coluna] = []
titanic_dados
```

```python
type(titanic_dados)
```

```python
# Verificando cada item e cada chave
linhas[1]
```

```python
# Aqui trataremos todas as linhas, excluindo a linha 0, que é o cabeçalho
for dado in linhas[1:]:
    dado_tratado = tratar_nome(dado)
    print(dado_tratado)
```

```python
type(dado_tratado)
```

```python
# Aqui faremos a mesma coisa, porém vamos separar as linhas pela vírgula e converter em uma lista!
for dado in linhas[1:]:
    dado_tratado = tratar_nome(dado)
    dado_como_lista = dado_tratado.split(',')
    print(dado_como_lista)
```
